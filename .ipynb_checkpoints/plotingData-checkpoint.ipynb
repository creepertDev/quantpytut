```python
import datetime as dt
import pandas as pd
import yfinance as yf
import plotly.offline as pyo
import plotly.graph_objects as gp
from plotly.subplots import make_subplots

pyo.init_notebook_mode(connected = True)
```

```python
end = dt.datetime.now()
start = dt.datetime(2025,1,1)

df = yf.download("QQQ", start, end)
print(df.head())
```

```python
df['MA50'] = df['Close'].rolling(window = 10, min_periods = 0).mean()
df['MA50'].head(20)
```

```python
df['MA50'] = df['Close'].rolling(window=10, min_periods=0).mean()
print(df['MA50'].head(20))
```

```python
fig = make_subplots(rows = 2, cols = 1, shared_xaxes = True, vertical_spacing = 0.1, subplot_titles = ('CBA', 'Volume'), row_width = [0.2,0.7])
```

```python
fig.add_trace(
    gp.Candlestick(x =df.index, open = df['Open'], high = df['High'], low = df['Low'], close = df['Close'], name = 'OHLC')
)
```
